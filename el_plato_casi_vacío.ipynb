# **El Plato  Casi Vacío**
## Análisis del poder adquisitivo como motor de la emigración

Detrás de cada estadística de migración hay una mesa familiar, que es el lugar donde se comparten los sueños y se discuten las ausencias. Sin embargo, en los últimos años, esa mesa ha comenzado a hablarnos de otra cosa: de una aritmética que no cierra y de un plato que, a pesar del esfuerzo, permanece casi vacío.

Cuba ha experimentado uno de los flujos migratorios más grandes de su historia hacia países como **Brasil, México, Uruguay y Estados Unidos**. Aunque las razones son múltiples, la economía doméstica es, sin duda, el motor principal.

Este proyecto se propone responder una pregunta fundamental: ¿Es el salario estatal cubano suficiente para cubrir una alimentación básica? analizando no solamente números, sino la distancia que existe justamente entre un salario estatal y el costo de la vida, una brecha que para muchos se ha vuelto tan ancha que solo se puede cruzar con un boleto de avión.



### **Inventario de la Escacez**: Anatomía de la Canasta Básica

Antes de hablar de fronteras, debemos hablar de lo que queda en casa. La canasta básica no es solo una lista de víveres; es el mapa de lo que una persona necesita para sostener el día a día. Para entender por qué el plato se siente vacío, primero debemos desglosar su contenido, por ello hemos recolectado los precios de los productos esenciales en los nuevos comercios (MIPYMES).

Debemos entender el costo de la vida en su expresión más cruda antes de analizar los ingresos. He construido esta canasta básica de forma personal, seleccionando únicamente aquellos productos indispensables para la subsistencia, la cual tiene un costo de 16 170 pesos cubanos (36.75 usd). No hay lujos en esta lista, ni caprichos; es el inventario de lo que un ser humano necesita para sostener el cuerpo y enfrentar el día.

```python
import graficos as gr
import treecko as tk

gr.pastel(tk.gasto,umbral=5)

```

### La pregunta inevitable

Sabemos ahora el precio de lo básico, no estamos hablando de lujos, ni siquiera de una dieta equilibrada; estamos hablando del inventario del superviviente. Pero los precios, por sí solos, son solo la mitad de la tragedia. La otra mitad reside en el bolsillo de quien trabaja.

Si esto es lo mínimo que necesitamos para que el plato no esté vacío, surge una pregunta que es, en realidad, el motor silencioso de todo éxodo: ¿Es el salario estatal cubano capaz de cubrir estos gastos elementales?

### **El Valor del Tiempo** : Salarios que Desvanecen

Pasamos del estante de la tienda a la billetera del trabajador. El salario estatal es, en teoría, la recompensa por el tiempo, el talento y el sacrificio de profesionales que sostienen los pilares de la sociedad.

Sin embargo, en el contexto del mercado y la inflación actual, ese tiempo parece perder su valor. Aquí analizaremos cómo se comporta el salario frente a la canasta de supervivencia que definimos anteriormente. Veremos que no se trata de una falta de esfuerzo, sino de una aritmética cruel donde el 100% de una vida de trabajo a veces no alcanza para cubrir el 100% de un plato básico.


```python
gr.percent(tk.precio_canasta,tk.salarios)
```

#### La Anatomía del Déficit

El gráfico anterior no es solo una estadística; es una bofetada de realidad. Cuando vemos que la barra de un médico o de un educador supera el **100%**, entramos en el terreno de lo imposible. 

Lo que estos datos nos gritan es que el salario estatal ha dejado de ser una herramienta de progreso para convertirse en un símbolo de impotencia. Nos enfrentamos a tres verdades dolorosas que explican parte del éxodo:

1. **La Paradoja del Talento:** Quienes más años dedicaron a su formación (médicos, maestros, ingenieros) son precisamente quienes ven cómo su sueldo se disuelve antes de llegar a la mitad del mes. El sistema les exige su máximo esfuerzo, pero les devuelve menos de lo mínimo para sobrevivir.

2. **El Sacrificio de lo Secundario:** Si el 100% (o más) del salario debe ir destinado a la canasta de supervivencia que yo mismo definí, ¿qué queda para lo demás? Se sacrifica la ropa, el calzado, el mantenimiento del hogar y la medicina. Se vive en un estado de emergencia perpetua donde el futuro se cancela para poder pagar el presente.

3. **La Ruptura del Contrato Social:** Un ciudadano trabaja para el Estado bajo la promesa de que su labor le garantizará una vida digna. Cuando el trabajo a tiempo completo no garantiza ni el plato básico, el contrato se rompe. El trabajador no elige irse porque quiera abandonar su vocación, sino porque su vocación ya no puede alimentarlo.

Este déficit no es solo económico, es emocional. Es la angustia de saber que, incluso trabajando sin descanso, la cuenta siempre da negativo. Por eso, el horizonte de **México, Uruguay o Estados Unidos** deja de verse como una aventura y empieza a verse como la única salida aritmética a una vida que no permite sumas, solo restas.

### **De la Abstraccíon al Mostrador** : El Sudor de las Unidades

Hasta ahora, hemos visto números que asustan: porcentajes que superan el 100%, el 200% y hasta el 300% de un ingreso mensual. Pero, ¿qué significan esas cifras cuando llegamos frente al mostrador de la tienda?

Para entender la verdadera profundidad del abismo, debemos traducir el salario estatal a algo tangible. No hablemos de porcentajes, hablemos de unidades. Si un trabajador decidiera dedicar todo su esfuerzo de un mes a comprar un solo producto esencial, ¿cuántos de ellos podría cargar en sus manos?

```python
gr.unidades_salarios(2100)
```



Este gráfico es la representación visual de lo que llamamos impotencia adquisitiva. Al observar los números que coronan cada barra, la realidad se vuelve indiscutible: lo que antes eran treinta días de vida y trabajo, hoy se traducen en apenas un puñado de artículos.

¿Qué nos revela este análisis?

* *La barrera de la unidad* : Es alarmante observar cuántos productos tienen barras cercanas al valor de 1. Esto significa que un profesional debe trabajar un mes entero para comprar, por ejemplo, apenas dos litros de aceite o un kilogramo de carne, sin que le quede un solo centavo para nada más, o tan siquiera poder comprar un cartón de huevos

* *La moneda del sacrificio* : Cuando la recompensa por un mes de labor se cuenta con los dedos de una mano, el trabajo pierde su función de progreso y se convierte en un simple mecanismo de agotamiento.

* *El motor del adiós* : Esta gráfica explica por qué la mirada del cubano se vuelve hacia el horizonte. No se emigra solo por ambición; se emigra porque la relación entre 'tiempo trabajado' y 'alimento obtenido' se ha roto. En este gráfico se entiende que, en Cuba, el sudor de un mes ya no alcanza para llenar siquiera una bolsa pequeña de mercado.

Es en esta desproporción donde el 'Plato Casi Vacío' deja de ser una metáfora para convertirse en una certeza física.

### **Geografías de la Esperanza**: El Horizonte Internacional

Después de observar cómo el salario en Cuba se rinde ante el costo de la vida, la pregunta deja de ser "¿por qué se van?" y se convierte en "¿hacia dónde es posible vivir?". El mapa de la emigración cubana no se dibuja al azar; se traza siguiendo la ruta de aquellos países donde el trabajo recupera su sentido original: la capacidad de proveer.

En esta última etapa de nuestro análisis, comparamos la realidad cubana con la de los destinos más buscados en Ámerica: **Brasil, México, Uruguay y Estados Unidos**. 

Lo que veremos a continuación es el contraste entre dos mundos. Mientras que en Cuba la canasta de supervivencia es un gigante que devora varios salarios, en estos países la relación se invierte. Allí, la comida —esa misma base indispensable que definimos al inicio— representa solo una fracción del ingreso mínimo. 

La emigración, vista a través de estos datos, no es una huida hacia el lujo, sino una búsqueda desesperada de **lógica aritmética**. Es el camino hacia un horizonte donde el plato no sea el fin del camino, sino el punto de partida para poder construir todo lo demás.

## **Conclusiones: La Aritmética del Adiós**

Tras recorrer el camino de los datos, desde el costo de un grano de arroz hasta la comparativa salarial entre naciones, las cifras nos entregan un veredicto claro. Este análisis no solo confirma una crisis económica, sino que documenta la fractura del proyecto de vida dentro de la isla.

#### **1. El Trabajo como Esfuerzo sin Retorno**
Nuestra investigación demostró que el salario estatal cubano ha perdido su función más elemental: la de garantizar la supervivencia. Cuando un profesional de la salud o la educación necesita más del **200% o 300%** de su ingreso para cubrir una canasta mínima diseñada para no morir de hambre, el trabajo deja de ser un derecho para convertirse en una condena al déficit permanente.

#### **2. La Desproporción de la Unidad**
Al traducir el salario a unidades físicas, el impacto es demoledor. Un mes de vida laboral equivale a apenas un par de artículos básicos. Esta "moneda del sacrificio" es la que realmente mide la crisis: un país donde el tiempo de un hombre o una mujer no alcanza para comprar siquiera el alimento de una semana, es un país que está empujando a sus ciudadanos hacia afuera.

#### **3. La Emigración como Única Respuesta Lógica**
La comparativa internacional con **Brasil, México, Uruguay y Estados Unidos** nos reveló que la emigración no es una búsqueda de opulencia, sino de normalidad. Mientras que en Cuba la supervivencia es un gigante inalcanzable, en los países de destino el trabajo recupera su propósito: alimentar al que lo realiza y permitirle soñar con el día siguiente.
